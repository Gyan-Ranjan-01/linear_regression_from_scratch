```python
x["social_media_hours"] = x["social_media_hours"].fillna(x["social_media_hours"].median())
```

```python
x["focus_score"] = x["focus_score"].fillna(x["focus_score"].mode()[0])
```

```python
x["sleep_hours"] = x["sleep_hours"].fillna(x["sleep_hours"].mean())
```

```python
print(x.shape)
```

```python
y = np.array(x['final_exam_score'].values)
x= np.array(x.drop(['final_exam_score','performance_category','student_id','career_goal','learning_style','mental_state','city_type','gender'],axis=1).values)
```

```python
#standardisation
std = x.std(axis=0)
std[std==0] = 1
x = (x - x.mean(axis=0)) / std
```

```python
n_features = x.shape[1]
print(n_features)
n_sources = x.shape[0]
```

```python
coeff = np.zeros(n_features)
coeff.shape
```

Method 1: nested loops

```python
#the method I have used is too basic, slow and vectorization can increase the speed of execution by nearly 1000 times

a=0.005
c=0
for i in range(1000):
    sume=0
    sumxe=np.zeros(n_features)
    yi = x @ coeff 
    yi += c
    for j in range(yi.shape[0]):
        e = y[j]-yi[j]
        for k in range(n_features):
            sumxe[k]+=x[j][k]*e
        sume+=e
    for j in range(n_features):
        coeff[j] = coeff[j] - (a)*((-2/x.shape[0])*sumxe[j])
    c -= (a)*((-1/x.shape[0])*sume)
```

Method 2: Vectorisation Method

```python
c=0
a = 0.005
for i in range(1000):
    yi = x @ coeff + c
    e = y - yi
    sumxe = x.T @ e
    sume = e.sum()
    coeff -= (-2/n_sources)*a*sumxe
    c -= (-1/n_sources)*a*sume
```

```python
for i in range(n_features):
    print(f'm{i+1} = {coeff[i]}')
print(f'c = {c}')
```
